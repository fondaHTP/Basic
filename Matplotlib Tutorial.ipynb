# MATPLOTLIB

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Tentukan Nilai x
x = np.arange(0,10)
```

```python
#Tentukan nilai 
y1 = x*10
```

```python
#Tentukan nilai y2
y2 = x**2+10 
```

```python
x
```

```python
y1
```

```python
y2
```

```python
plt.plot(x,y1,'b--',label = 'x*10')
plt.legend()
```

```python
plt.plot(x,y2,'r',label='x**2+2 ')
plt.legend()
```

```python
#buat label x-Axis dan y-Axix
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
#tentukan grafik y1 dengan garis biru putus putus beserta label
plt.plot(x,y1, 'b--', label = 'x*10')
#tentukan grafik y2 dengan garis merah beserta label
plt.plot(x,y2,'r',label='x**2+10 ')
#buat legend
plt.legend()
# Buat header grafik
plt.title('Grafik Matplotlib', fontdict={'fontname': 'Arial', 'fontsize': 24})
#Simpan hasil grafik
plt.savefig('matplotlib.png', dpi=300)
```

#  Grafik Bar dengan Data
