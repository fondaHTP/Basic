```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10)
```

```python
y1=x*10
```

```python
y2=x**2+10
```

```python
x
```

```python
y1
```

```python
y2
```

```python
plt.plot(x,y1,'--',label='x*10')
plt.legend()
```

```python
plt.plot(x,y2,'r', label='x**2+10')
plt.legend()
```

```python
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(x,y1,'--',label='x*10')
plt.plot(x,y2,'r', label='x**2+10')
plt.legend()
plt.title('Grafik Matplotlib', fontdict={'fontname': 'Arial', 'fontsize':24})
plt.savefig('matpotlibfirst.png', dpi=300)
```
